# 环境及数据准备

## 环境及依赖

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
#import openpyxl
```

## 数据读入

```python
origin_data = open("/home/xavier/Downloads/Data/data.txt", "r", encoding='utf-8')
#因为不是通过pandas读入文件，我们此处不能使用head函数
#读取一行数据
line = origin_data.readline()
line
```

# 数据处理

## 获取列名集

```python
# 创建空列表
title_list = []
# 迭代所有行，直到把文件所有行读完
while line:
    # 去掉换行符
    line = line.strip()
    # 把一行数据按照“，”隔开为一个个商品名
    product_items = line.split(',')
    # 将这些商品名都加入到列表中
    for item in product_items:
        title_list.append(item)
    # 读取下一行，继续迭代
    line = origin_data.readline()
# 使用set数据结构去重
title_list = set(title_list)
title_list
```

```python
# 打印商品列表长度
print("共有", len(title_list) ,"种商品")
```

```python
# 关闭文件
origin_data.close()
```

## 数据转换

```python
# 创建一个空的dataframe用来存放处理后的数据集
changed_data = pd.DataFrame(columns=list(title_list))
# 打开文件
origin_data = open("/home/xavier/Downloads/Data/data.txt", "r", encoding='utf-8')
```

```python
# 读取一行数据
line = origin_data.readline()
# 迭代所有行，直到把文件所有行读完
while line:
    # 去掉换行符
    line = line.strip()
    # 存储这条记录中，各个商品的购买情况
    customer = {}
    # 把一行数据根据“，”隔开为一个个商品名
    product_items = line.split(',')
    # 逐个查看是否购买商品
    for item in title_list:
        if(product_items.__contains__(item)) :
            customer[item] = 1
        else:
            customer[item] = 0
    # 将这一行添加到数据集
    changed_data = pd.concat([changed_data, pd.DataFrame.from_records([customer])])
    # 读取下一行，继续迭代
    line = origin_data.readline()
# 打印前五行数据
changed_data.head(5)
```

```python
# 将处理后的数据集保存
changed_data.to_excel("/home/xavier/Downloads/Data/changed_data.xlsx", index=False)
```

# 关联数据挖掘

```python
# 将所有列转换为bool类型
changed_data = changed_data.astype(bool)
```

```python
frequent_itemsets = apriori(changed_data, min_support=0.03, use_colnames=True)
# 输出频繁项集
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
